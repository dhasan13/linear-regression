```python
import pandas as pd

# Create dataset
data = {
    'year': [1970, 1971, 1972, 1973, 1974, 1975, 1976, 1977],
    'income': [3399, 3768, 4251, 4804, 5576, 5998, 7062, 7100]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Convert income into 0 or 1
# 1 = High income, 0 = Low income
df['income_category'] = df['income'].apply(lambda x: 1 if x > 5000 else 0)

print(df)
```

```python
X = df[['year']]
y = df['income_category']
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, y)

predictions = model.predict(X)

print("Predictions:", predictions)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.xlabel("Year")
plt.ylabel("Income Category (0/1)")
plt.title("Logistic Regression Graph")
plt.show()
```
